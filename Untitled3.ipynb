```python
# 📌 Concept: Linear Regression (Basic Machine Learning)

# Linear Regression is used to predict a value based on the relationship between variables.

# 🎯 Example:

# Predict house prices based on house size.

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data (House size in square feet)
X = np.array([500, 1000, 1500, 2000, 2500]).reshape(-1, 1)

# House prices
y = np.array([150000, 300000, 450000, 600000, 750000])

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict price for 1800 sq ft
predicted_price = model.predict([[1800]])

print("Predicted price for 1800 sq ft:", predicted_price[0])

# Plot data
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel("House Size (sq ft)")
plt.ylabel("Price")
plt.title("Linear Regression Example")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create dataset
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Math": [85, 78, 92, 70, 88],
    "Science": [90, 75, 85, 80, 95],
    "English": [88, 82, 90, 72, 85]
}

df = pd.DataFrame(data)

# Step 2: Calculate Total and Average
df["Total"] = df["Math"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3

print("Student Data:\n")
print(df)

# Step 3: Find Top Scorer
top_student = df.loc[df["Total"].idxmax()]
print("\nTop Scorer:")
print(top_student["Name"], "with total marks:", top_student["Total"])

# Step 4: Visualization
plt.bar(df["Name"], df["Total"], color='skyblue')
plt.xlabel("Students")
plt.ylabel("Total Marks")
plt.title("Student Total Marks")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create Sales Dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [2500, 3000, 2800, 3500, 4000, 3700]
}

df = pd.DataFrame(data)

# Step 2: Basic Analysis
total_sales = df["Sales"].sum()
average_sales = df["Sales"].mean()
highest_month = df.loc[df["Sales"].idxmax()]
lowest_month = df.loc[df["Sales"].idxmin()]

print("Sales Data:\n")
print(df)

print("\nTotal Sales:", total_sales)
print("Average Sales:", average_sales)
print("Highest Sales Month:", highest_month["Month"], "-", highest_month["Sales"])
print("Lowest Sales Month:", lowest_month["Month"], "-", lowest_month["Sales"])

# Step 3: Visualization
plt.plot(df["Month"], df["Sales"], marker='o', color='green')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.grid(True)
plt.show()
```
